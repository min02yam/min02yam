## 3-1. 세종 테이블 결측 대체

2. 원인 세종이 2016부터 있음. 이거 때문인듯 그 전값은 충남의 값으로 대체

```r
final = fread("ASOS_예보_최종합.csv")
```

```r
final %>% summary
dt_area $area %>% table
```

```r
#step1. 충남만(2012-01-01 ~ 2015-12-31) 따로 빼기 - 이어 붙일 준비
dt_area %>% filter( area == '충남') -> dt_충남

which = which(dt_충남$yyyymmdd== '2015-12-31')
dt_충남_slice = slice(dt_충남, 1: which )
dt_충남_slice %<>% mutate(area = gsub("충남","세종",area))
```

```r
dt_충남_slice %<>% mutate(area = gsub("충남","세종",area))
```

```r
#step2. dt_area에 rbind 하기
air = rbind(dt_area, dt_충남_slice) 
air %<>% arrange(yyyymmdd, area)
```

```r
# 확인
air$area %>% table 
```

```r
air %>% summary
```

## 3-2.성별 변수 만들기

```r
성별 = rbind(air, air)
```

```r
성별 %>%
  arrange(yyyymmdd, area) %>% 
  mutate(sex = rep(1:2, times=nrow(성별)/2 )) -> ready
```

## 3-3. 최종 테이블 완성

```r
#키변수 형식 변경 
final %<>% mutate(yyyymmdd=parse_date_time(yyyymmdd, orders="ymd"))
```

```r
new =  final  %>% right_join(ready, by=c("yyyymmdd", "area","sex"))
```

```r
write.csv(new, "ASOS_예보_대기_최종합.csv", row.names=FALSE)
```
